```python
import json
import networkx as nx
from networkx.readwrite import json_graph
import matplotlib.pyplot as plt
```

```python
test = json.load(open('example.json', 'r'))

G = json_graph.node_link_graph(test)
nx.draw(G)
plt.show()
```

```python
# -------- desired output format --------
# {'data': {'id': 'one', 'label': 'Node 1'}},
# {'data': {'id': 'two', 'label': 'Node 2'}},
# {'data': {'source': 'one', 'target': 'two', 'label': 'Node 1 to 2'}}]

# create nodes and links
nodes = [{'data': node } for node in test['nodes']]
links = [{'data': link } for link in test['links']]

# add label name
for key in nodes:
    name = key['data']['id']
    #print(name)
    key['data']['label'] = name

for key in links:
    src = key['data']['source']; tg = key['data']['target']
    #print((src,tg))
    key['data']['label'] = 'Node %s to %s' % (src,tg)

# combine
both = nodes + links
```

```python
from jupyter_dash import JupyterDash
import dash_cytoscape as cyto
import dash_html_components as html

app = JupyterDash(__name__)

app.layout = html.Div([
    cyto.Cytoscape(
        id='cytoscape-display-nodes',
        layout={'name': 'breadthfirst'},
        style={'width': '70%', 'height': '400px'},
        elements=both
    )
])

app.run_server(debug=True,port=8053)
```
